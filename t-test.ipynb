```python
import numpy as np
from scipy import stats
```

```python
sample =[172,174,168,169,171,173,175,170,169,172]
samp_mean = np.mean(sample)
samp_std = np.std(sample,ddof = 1)
n = len(sample)
mean_population = 170
```

```python
samp_mean = np.mean(sample)
samp_mean
```

```python
t_score = (samp_mean - mean_population)/ (samp_std/ np.sqrt(n))
```

```python
t_score
```

```python
p_value = 2*(1-stats.t.cdf(abs(t_score),df = n-1))
```

```python
p_value
```

```python
alpha = 0.05
```

```python
if p_value < alpha:
    print("reject the null hypothesis")
else:
    print("accept the null hypothesis")
```
