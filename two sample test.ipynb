```python
import numpy as np
from scipy import stats
```

```python
grp_a = [85,88,90,92,87,85,89,91,86,88]
grp_b = [82,84,80,83,81,79,78,85,84,83]
```

```python
t_stats, p_value = stats.ttest_ind(grp_a, grp_b, equal_var=False)
```

```python
t_stats
```

```python
p_value
```

```python
alpha = 0.05
```

```python
if p_value< alpha:
    print("i will reject the null hypothesis")
else:
    print("i will accept the null hypothesis")
```
